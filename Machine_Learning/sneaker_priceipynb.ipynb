```python
import matplotlib.pyplot as plt

# Count occurrences of each brand
brand_counts = df['Brand'].value_counts()

# Plot
plt.figure(figsize=(20, 10))  # Wide figure for better readability
plt.bar(brand_counts.index, brand_counts.values)

# Labels and title
plt.xlabel('Brand')
plt.ylabel('Number of Sales')
plt.title('Sales Count by Brand')

# Rotate brand names for better visibility
plt.xticks(rotation=45, ha='right')

# Optional: Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# Count occurrences of each brand
brand_counts = df['Title'].value_counts()

# Plot
plt.figure(figsize=(100, 100))  # Wide figure for better readability
plt.bar(brand_counts.index, brand_counts.values)

# Labels and title
plt.xlabel('Title')
plt.ylabel('Number')
plt.title('Sales Count by Title')

# Rotate brand names for better visibility
plt.xticks(rotation=45, ha='right')

# Optional: Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
