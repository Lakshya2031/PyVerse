```python
import requests

def get_lat_lng_from_pincode(pincode, api_key):
    """Convert pin code to latitude & longitude using Geocoding API."""
    geocode_url = f"https://maps.googleapis.com/maps/api/geocode/json?address={pincode}&key={api_key}"
    response = requests.get(geocode_url)
    data = response.json()

    if data["status"] == "OK":
        location = data["results"][0]["geometry"]["location"]
        return location["lat"], location["lng"]
    else:
        return None, None

def get_nearby_hospitals(lat, lng, api_key, radius=5000):
    """Fetch nearby hospitals using Google Places API."""
    url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    params = {
        "location": f"{lat},{lng}",
        "radius": radius,
        "type": "hospital",
        "key": api_key
    }

    response = requests.get(url, params=params)
    results = response.json().get("results", [])

    hospitals = []
    for place in results:
        name = place.get("name")
        address = place.get("vicinity")
        hospitals.append(f"{name} - {address}")

    return hospitals
```
